### Probability mass function (PMF
* For discrete data
* Finds all the values that a random value might take

<p style="text-align: center;font-weight: 900px;font-size: 24px;">Coin toss (two coins)</p>

```python
n_tosses = 2
res = np.random.randint(0, 2, n_tosses)
print(res)

# If heads == 1 and tails == 0, get number of heads in sequence.
n_heads = res.sum()
print(f"Number of heads results for {n_tosses} tosses: {n_heads}")
```

#### Larger scale

```python
n_epochs = 1000
heads = np.zeros(n_epochs, dtype = np.uint32)
for i in range(n_epochs):
    heads[i] = np.random.randint(0, 2, 2).sum()
plt.stem(np.bincount(heads), use_line_collection=True)
```
