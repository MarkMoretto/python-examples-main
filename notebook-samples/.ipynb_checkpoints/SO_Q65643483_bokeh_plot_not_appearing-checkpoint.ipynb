```python
import re
import pandas as pd
from bokeh.plotting import figure, show
from bokeh.io import output_notebook
from bokeh.models.tools import HoverTool
```

```python
sample = """
2018-10-22  7468.629883 2.282400e+09 0.263123    NASDAQ
2018-10-23  7437.540039 2.735820e+09    -0.416272   NASDAQ
2018-10-24  7108.399902 2.935550e+09    -4.425390   NASDAQ
2018-10-25  7318.339844 2.741810e+09    2.953406    NASDAQ
2018-10-26  7167.209961 2.964780e+09    -2.065084   NASDAQ
""".strip()


lines = [re.split(r"\s+", line) for line in sample.split("\n")]
df = pd.DataFrame(data=lines)
df.columns = ["Date","Adj Close","Volume","Day_Perc_Change","Name"]




df.loc[: , "Date"] = pd.to_datetime(df.loc[: , "Date"], infer_datetime_format = True)
df.loc[: , "Adj Close"] = df.loc[: , "Adj Close"].astype(float)
print(df)
```

```python
output_notebook()
```

```python
# Create HoverTool class object with appropriate formatting for the desired tooltip output.
# More in the Bokeh inspectors section: https://docs.bokeh.org/en/latest/docs/user_guide/tools.html#inspectors
ht = HoverTool(
    tooltips = [
        # Formatting similar to strftime().
        # Docs: https://docs.bokeh.org/en/latest/docs/reference/models/formatters.html#bokeh.models.formatters.DatetimeTickFormatter
        ("Date", "@Date{%Y-%m-%d}"),
        
        # Format to two decimal ponts.  We also have the Hong Kong dollar sign (HK$) at the start.
        ("Adj Close", "HK$@{Adj Close}{%0.2f}"),
    ],
    formatters = {
        "@Date": "datetime",
        "@{Adj Close}": "printf",
    }
)


# s = figure(plot_height=600, plot_width=1000, title = "Adjacent Closing Prices", x_axis_label='Date', y_axis_label='Price')
s = figure(
    plot_height = 450,
    plot_width = 600,
    title = "Adjacent Closing Prices",
    x_axis_label = "Date",
    y_axis_label = "Price",
    x_axis_type = "datetime", # For x-axis datetime formatting (obviously)
)

# Add your hovertools object here.
s.add_tools(ht)

s.line(source=df, x = "Date", y = "Adj Close", line_width = 2)

show(s)
```
